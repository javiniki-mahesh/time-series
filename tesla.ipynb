### Time resampling

Reference : https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78#:~:text=Resampling%20is%20used%20in%20time,DatetimeIndex%2C%20PeriodIndex%2C%20or%20TimedeltaIndex.

There are several time resampling techniques available in pandas:

    Downsampling: Downsampling refers to the process of aggregating data from a higher frequency to a lower frequency. This is done by grouping data over a longer time interval, such as from hourly to daily data.

    Upsampling: Upsampling is the opposite of downsampling, and refers to the process of increasing the frequency of a time series. This is done by introducing additional data points in the time series, typically by using interpolation techniques.

    Rolling window: A rolling window is a technique for calculating statistics on a time series by sliding a window of fixed size over the data. This is useful for smoothing the data and identifying trends.

    Expanding window: An expanding window is a technique that calculates statistics over an increasingly larger portion of the time series. This can be useful for identifying long-term trends in the data.

    Resampling with custom frequencies: Pandas also allows you to resample time series data using custom frequencies, such as every 2 weeks or every 6 months. This can be useful if you need to analyze data at a specific frequency that is not available by default.

    Overall, the choice of resampling technique depends on the specific problem you are trying to solve and the characteristics of your time series data.

### Downsampling

```python
# min annual frequency
tesla.resample(rule = 'A').min()
```

```python
# max annual frequency
tesla.resample(rule = 'A').max()
```

The rule parameter in the resample() method of pandas specifies the frequency at which the data will be resampled. It can take different values depending on the desired frequency of the resampled data. Some of the most commonly used frequency values for the rule parameter are:

    'D' or 'day': resample to daily frequency.
    'W' or 'week': resample to weekly frequency.
    'M' or 'month': resample to end-of-month frequency.
    'Q' or 'quarter': resample to quarterly frequency.
    'Y' or 'year': resample to annual frequency.
    
In addition to these, pandas also allows custom date offsets to be specified using the DateOffset class, such as '2W' for every 2 weeks or '6M' for every 6 months.

The rule parameter can also take a combination of these frequency values and date offsets, such as '2D' for every 2 days or '3M1W' for every 3 months and 1 week.

Overall, the choice of rule depends on the specific problem you are trying to solve and the characteristics of your time series data.
